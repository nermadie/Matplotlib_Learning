## 1. Pyplot API vs Object-Oriented API

- Pyplot API -> quick and easy plots
- Object-Oriented API -> Advanced plots


### 1.1. Pyplot API


```python
xmin = 0
xmax = 4
ymin = -0.6
ymax = 0.6
```

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), color="blue", linestyle="dashed", label="Sin(X)")
plt.plot(x, np.cos(x), color="red", label="Cos(X)")

plt.title("Sin and Cos Functions")  # Title of the plot
plt.xlabel("X")  # Label the x-axis
plt.ylabel("F(X)")  # Label the y-axis

# plt.xlim([0, 4]) # Limit the x-axis
# plt.ylim([-0.6, 0.6]) # Limit the y-axis
# plt.axis([xmin, xmax, ymin, ymax])  # Limit the x and y axis
# plt.axis("tight")  # Tighten the axis to the plot
plt.axis("equal")  # Make the axis equal in length


plt.legend()  # Show the legend (descriptive information about the plot) -> label parameter is required in the plot function
# plt.show() # Only required when running the script in the terminal (not in Jupyter Notebook)
```

### 1.2. Object-Oriented API


```python
# 1. Prepare data
x = np.arange(1, 5)
y = np.arange(11, 45, 11)

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5))  # Figure size = Width & Height of the plot

# 3. Plot data
ax.plot(x, y)

ax.set(title="Simple Plot", xlabel="X-Axis", ylabel="Y-Axis")
```

```python
import numpy as np

# Setting the seed for the global random number generator
np.random.seed(42)
print(np.random.rand())  # Global RNG

# Creating a separate random number generator instance
rng = np.random.RandomState(42)
print(rng.rand())  # Independent RNG instance

# The global state is not affected by rng
print(np.random.rand())
```

```python
# OO API
# Prepare New Data
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x, y, s=sizes, c=colors, cmap="viridis", alpha=0.3)

fig.colorbar(img1)
```

## 2.3. Bar

- Vertical
- Horizontal


```python
# Prepare Data from a python dictionary
soft_drink_prices = {"Coke": 10, "Pepsi": 12, "Sprite": 8}

fig, ax = plt.subplots()

# Vertical Bar Chart
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(
    title="Bach Hoa Xanh's Soft Drink Prices", ylabel="Price ($)", xlabel="Soft Drink"
)
```

```python
fig, ax = plt.subplots()

# Horizontal Bar Chart
# ** Note: must use list() to convert the dict_keys object to a list
ax.barh(list(soft_drink_prices.keys()), soft_drink_prices.values())
ax.set(
    title="Bach Hoa Xanh's Soft Drink Prices", xlabel="Price ($)", ylabel="Soft Drink"
)
```

## 2.4. Histogram


```python
# Prepare Data
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
student_height[:10]
```

```python
# Pyplot API
plt.hist(student_height, bins=30)  # bins = number of bars in the histogram
```

```python
# OO API
fig, ax = plt.subplots()
ax.hist(student_height)
```

## 2.5. Subplot

- Multiple plots in the same figure: [Matplotlib docs](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html)


```python
# Option 1: get each axis separately
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot data to each axis
ax1.plot(x, x / 2)
ax1.plot(x, x * 2, color="red")

ax2.scatter(np.random.random(10), np.random.random(10))

ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values())

ax4.hist(student_height, bins=30)
```

```python
# Option 2: get all axes at once
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot data to each axis
ax[0, 0].plot(x, x / 2)
ax[0, 0].plot(x, x * 2, color="red")

ax[0, 1].scatter(np.random.random(10), np.random.random(10))

ax[1, 0].bar(soft_drink_prices.keys(), soft_drink_prices.values())

ax[1, 1].hist(student_height, bins=30)
```

## 3. Project: Data Visualization - California Cities


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv("california_cities.csv", index_col=0)
cities
```

```python
# Extract latd (vi do) and longd (kinh do) from the DataFrame
lat, lon = cities["latd"], cities["longd"]
population, area = cities["population_total"], cities["area_total_km2"]

cities["area_land_km2"].sort_values(ascending=False)
```

```python
plt.style.use("seaborn-v0_8")
plt.figure(figsize=(8, 6))

# Plot using the Pyplot API
plt.scatter(
    lon, lat, c=np.log10(population), cmap="viridis", s=area, linewidths=0, alpha=0.5
)

plt.axis("equal")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(label="log$_{10}$(population)")
plt.clim(3, 7)

# Create a legend for cities's size
area_range = [50, 100, 300, 500]
for area in area_range:
    plt.scatter([], [], s=area, c="k", label=str(area) + " km$^2$", alpha=0.5)
plt.legend(
    labelspacing=1,
    title="City Area",
)

plt.title("California Cities: Population and Area Distribution")
```
